```python
%matplotlib inline

from __future__ import division, absolute_import

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import os
```

```python


# Import MNIST data
# from tensorflow.examples.tutorials.mnist import input_data
# mnist = input_data.read_data_sets("MNIST_data", one_hot=True)

def norm_df(df):
    df = df  / df.values.max()
    return df

final_df = pd.DataFrame()
cols = []
dirFiles = '/Users/Nandita/Downloads/Archive/'
for file in os.listdir(dirFiles):
    if(file!='.DS_Store'):
        filename = dirFiles +  file
        #print filename
        df = pd.read_csv(dirFiles +  file, delimiter='\t')
        cols = df.ix[:,3]
        df = df.ix[:,10]
        final_df = pd.concat([final_df, df], axis= 1)
final_df = final_df.T
final_df.columns = cols
final_df = norm_df(final_df)
#final_df = final_df.loc[:, (final_df != 0).any(axis=0)]

X_train, X_test, y_train, y_test = train_test_split(final_df, final_df, test_size=0.5, random_state=1001)
############INIT
N_INPUT = final_df.shape[0] # MNIST data input (img shape: 28*28)
LEARNING_RATE = 1
TRAINING_EPOCHS = 500
NUM_EXAMPLES = 3
DISPLAY_STEP = 1
examples_to_show = 10

N_HIDDEN = 500

FILENAME = 'data/out2.txt'

weights = {
  'encoder_h1': tf.Variable(tf.random_normal([N_INPUT, N_HIDDEN])),
  'decoder_h1': tf.Variable(tf.random_normal([N_HIDDEN, N_INPUT])),
}
biases = {
  'encoder_b1': tf.Variable(tf.random_normal([N_HIDDEN])),
  'decoder_b1': tf.Variable(tf.random_normal([N_INPUT])),
}

sess = tf.InteractiveSession()
    
######MAIN
def main(sess):
    
    X = tf.placeholder("float", [None, N_INPUT])
    encoder_op = encoder(X)
    decoder_op = decoder(encoder_op)

    y_pred = decoder_op
    y_true = X

    cost = tf.reduce_mean(tf.pow(y_true - y_pred, 2))
    optimizer = tf.train.RMSPropOptimizer(LEARNING_RATE).minimize(cost)
       
    init = tf.global_variables_initializer()

        
    sess.run(init)

    for epoch in range(TRAINING_EPOCHS):
        batch_xs = X_train
        _, c = sess.run([optimizer, cost], feed_dict={X: batch_xs})
        c_test = sess.run(cost, feed_dict={X:X_test})
        if epoch % DISPLAY_STEP == 0:
            plt.plot(epoch, c,'+r')
            plt.plot(epoch, c_test, '+b')
            print("Epoch:", '%04d' % (epoch+1), "{:.9f}".format(c))

print("Optimization Finished!")

# Building the encoder
def encoder(x):
    layer_1 = tf.nn.sigmoid(tf.add(tf.matmul(x, weights['encoder_h1']),
                         biases['encoder_b1']))
    return layer_1


# Building the decoder
def decoder(x):
    layer_1 = tf.nn.sigmoid(tf.add(tf.matmul(x, weights['decoder_h1']),
                         biases['decoder_b1']))
    return layer_1


if __name__ == "__main__":
    main(sess)
```
